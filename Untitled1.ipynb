```python
import yfinance as yf
import ta
from newscatcherapi import NewsCatcherApiClient
# Загрузить данные о курсе акций компании Apple
aapl = yf.Ticker("AAPL")
df = aapl.history(period="max")

# Загрузить новости о компании Apple за последние 30 дней
api = NewsCatcherApiClient(x_api_key="fRgxmznpVrDb1B_OVSdTtUMKRkK3p5NaOHY6rERuCac")
news = api.get_search(q='YANDEX', lang='en', sort_by='relevancy', page_size=100)

# Объединить заголовки новостей в один текст
text = ' '.join([n['title'] for n in news['articles']])

# Анализировать текст с помощью NLP
from textblob import TextBlob

blob = TextBlob(text)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Отобразить результаты анализа
print('Sentiment Analysis Results:')
print(f'Polarity: {polarity:.2f}')
print(f'Subjectivity: {subjectivity:.2f}')
```

```python
import yfinance as yf
import ta

# Загрузить данные о курсе акций компании Apple
aapl = yf.Ticker("YNDX")
df = aapl.history(period="max")

# Вычислить различные технические показатели
df['ema20'] = ta.trend.EMAIndicator(df['Close'], window=20).ema_indicator()
df['sma50'] = ta.trend.SMAIndicator(df['Close'], window=50).sma_indicator()
df['rsi14'] = ta.momentum.RSIIndicator(df['Close'], window=14).rsi()


# Отобразить график курса акций и технических показателей
df[['Close', 'ema20', 'sma50']].plot(figsize=(12,8))
plt.title('AAPL Technical Analysis')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

# Отобразить график RSI и MACD
df[['rsi14']].plot(figsize=(12,4))
plt.title('AAPL RSI')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.show()

df[['macd', 'macdsignal']].plot(figsize=(12,4))
plt.title('AAPL MACD')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.show()
```
