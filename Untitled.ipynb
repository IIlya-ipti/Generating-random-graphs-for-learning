```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from colormath.color_objects import sRGBColor, LabColor
from colormath.color_conversions import convert_color
import time

def get_image(path):
    # загрузите изображение
    return cv2.imread(path, cv2.COLOR_BGR2RGB)

def get_image2(path):
    return cv2.imread(path)

# norms    
def rgb_to_cielab(a):
    """
    a is a pixel with RGB coloring
    """
    a1,a2,a3 = a/255
    color1_rgb = sRGBColor(a1, a2, a3);
    color1_lab = convert_color(color1_rgb, LabColor);
    return color1_lab

def norm(first,second):
    first = rgb_to_cielab(first).get_value_tuple()
    second = rgb_to_cielab(second).get_value_tuple()
    sm = 0
    for i in range(len(first)):
        sm += abs(first[i] - second[i])
    return sm 

def scelet_points(mask):
    ls = []
    n1 = len(mask)
    for i in range(len(mask)): 
        n2= len(mask[i])
        for j in range(len(mask[i])):
            if mask[i][j] == 255:
                ls.append((i,j))# поменял (j,i) -> (j,i)
    return ls

def get_points_mask(mask):
    ind = 0
    ls = []
    n1 = len(mask)
    for i in range(len(mask)): 
        n2= len(mask[i])
        for j in range(len(mask[i])):
            if mask[i][j] == 255:
                ls.append((i,j,ind))# поменял (j,i) -> (j,i)
                ind += 1
    return ls

def draw_line(img,points, thickness=5, color=(0, 0, 255)):
    # Загрузка изображения
    #points = np.array([[10, 50], [400, 50], [200, 200], [50, 400]], np.int32)
    # Рисование линии
    cv2.polylines(img, [points], isClosed=False, color=color, thickness=thickness)

def cv_norm(x,y):
    #return abs(x) + abs(y)
    return (x**2 + y**2)**0.5;
    
def arg_min(points_array:list, points):
    arg = -1
    min_value = float("inf")
    for i in range(len(points_array)):
        norm_acual = cv_norm(points_array[i][0] - points[0], points_array[i][1] - points[1])
        if norm_acual < min_value:
            min_value = norm_acual
            arg = i
    return (arg,min_value)

def get_metric(first,second,img, norm):
    first_pix = img[first[0]][first[1]]
    second_pix = img[second[0]][second[1]]
    return norm(first_pix,second_pix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

import cv2
import numpy as np

def illumination_normalize(image):
    # read input image
    hh, ww = image.shape[:2]
    max_size = max(hh, ww)

    # illumination normalization
    ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

    # separate channels
    y, cr, cb = cv2.split(ycrcb)

    # get background using Gaussian blur with standard deviation 5 pixels for 300x300 size image
    sigma = int(5 * max_size / 300)
    gaussian = cv2.GaussianBlur(y, (0, 0), sigma, sigma)

    # subtract background from Y channel
    y_normalized = (y - gaussian + 100)

    # merge channels back
    ycrcb_normalized = cv2.merge([y_normalized, cr, cb])

    # convert back to BGR
    output = cv2.cvtColor(ycrcb_normalized, cv2.COLOR_YCrCb2BGR)

    ## save results
    #cv2.imwrite('processed_' + image_path, output)
    return output
```

```python
import cv2
import numpy as np

def normalize_image(img):
    # Нормализация яркости и контраста
    norm_img = cv2.normalize(img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX).astype(np.uint8)
    
    # Нормализация цветового баланса
    r, g, b = cv2.split(norm_img)
    r_mean, g_mean, b_mean = np.mean(r), np.mean(g), np.mean(b)
    k = (r_mean + g_mean + b_mean) / 3
    r, g, b = r / r_mean * k, g / g_mean * k, b / b_mean * k
    norm_img = cv2.merge((r, g, b)).astype(np.uint8)

    # Нормализация освещенности
    hsv = cv2.cvtColor(norm_img, cv2.COLOR_BGR2HSV)
    hsv[:,:,2] += 25
    norm_img = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR).astype(np.uint8)

    return norm_img
```

```python
def equalHist(img):
    # Высота и ширина матрицы серого изображения
    h, w = img.shape
         # Шаг 1: Рассчитать гистограмму в градациях серого
    grayHist = calcGrayHist(img)
         # Шаг 2: Рассчитать накопленную серую гистограмму
    zeroCumuMoment = np.zeros([256], np.uint32)
    for p in range(256):
        if p == 0:
            zeroCumuMoment[p] = grayHist[0]
        else:
            zeroCumuMoment[p] = zeroCumuMoment[p - 1] + grayHist[p]
         # Шаг 3: Получить соотношение отображения между уровнем серого на входе и уровнем серого на основе накопленной гистограммы серого
    outPut_q = np.zeros([256], np.uint8)
    cofficient = 256.0 / (h * w)
    for p in range(256):
        q = cofficient * float(zeroCumuMoment[p]) - 1
        if q >= 0:
            outPut_q[p] = math.floor(q)
        else:
            outPut_q[p] = 0
         # Четвертый шаг: получить выровненное изображение гистограммы
    equalHistImage = np.zeros(img.shape, np.uint8)
    for i in range(h):
        for j in range(w):
            equalHistImage[i][j] = outPut_q[img[i][j]]
    return equalHistImage
```
